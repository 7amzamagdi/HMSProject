## Show Heat Map

```python
df1=df.copy()

df1.drop(columns=['expert_consensus'], inplace=True)
```

```python
correlation_matrix = df1.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
category_counts = df.groupby('expert_consensus')['eeg_id'].sum().reset_index()

# Plotting
plt.bar(category_counts['expert_consensus'], category_counts['eeg_id'], color='skyblue')
plt.xlabel('expert_consensus')
plt.title('Bar Plot for Categorical Features')
plt.show()
```
