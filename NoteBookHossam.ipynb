```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data= pd.read_csv('./TheData/train.csv')
data.sample(2)
```

```python
for column in data.select_dtypes(include='number'):
    plt.figure(figsize=(8, 4))  # Set the figure size
    sns.histplot(data=data, x=column, kde=True)  # Create the histogram with KDE (Kernel Density Estimation)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
